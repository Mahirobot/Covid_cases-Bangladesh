```
url = 'https://www.worldometers.info/world-population/bangladesh-population/'
```

```
result = requests.get(url)
c = result.content
```

```
soup = BeautifulSoup(c)
```

```
summary = soup.find('div',{'class':'table-responsive', 'style':'clear:both; width:100%; max-width:500px '})

tables = summary.find_all('table')
```

```
id = list(range(1,65))
dist_name = []
population = []
data = []
rows = tables[0].findAll('tr')
for tr in rows:
  cols = tr.findAll('td')

  for td in cols:

    text = td.find(text=True)
    data.append(text)
x = 1
y = 2    
for i in range(0,210): 
  if i== x:   
    dist_name.append(data[i])
    x += 3
  if i == y:
    population.append(data[i])
    y += 3

print(dist_name)
print(population)
```

```
data = pd.DataFrame({
    'District': dist_name,
    'Population':population,
})
data.to_csv('/content/drive/My Drive/Colab Notebooks/Mini projects/BD Cloropleth Map/population.csv')
```
