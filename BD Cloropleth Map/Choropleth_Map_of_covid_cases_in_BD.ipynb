```
#5
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Mini projects/BD Cloropleth Map/cases.csv")
print(len(df))
data = df
```

```
#6
from difflib import get_close_matches 
  
def closeMatches(patterns, word): 
     matches = get_close_matches(word, patterns) 
     if matches:
        first = matches[0]
     else:
        first = word
     return first
     
for A in data['District_name']:
  if A not in Objectid_map:
    x = closeMatches(Objectid_map, A)
    #print(str(A) + "will replace " + str(x))
    for feature in bd_states['features']:
        if feature['properties']['NAME_3'] == x:
          feature['properties']['NAME_3'] = A
          

for A in data['District_name']:
  if A not in Objectid_map:
    x = closeMatches(Objectid_map, A)
    print(x) 
```

```
#9

## INTENSITY OF COVID INFECTION ##

data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Mini projects/BD Cloropleth Map/covid_data.csv')
fig = px.choropleth(data,locations = 'id',geojson=bd_states,color = 'cases')
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
